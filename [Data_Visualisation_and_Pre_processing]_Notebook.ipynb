## 1.1 Numpy Basics

```
import numpy as np
```

Numpy is an incredible library which can do all sorts of numerical computations efficiently. It contains the numpy arrays which are some of the most frequently used constructs in Data Science. Many libraries including Pandas build upon the numpy arrays.

You can think of numpy arrays as collections of any objects. 1-d arrays can be thought of as vectors while 2-d arrays can be thought of as matrices. Nevertheless, np arrays can extend beyond that, but we will not be using much of that for our workshop.

### 1.1.1 Creating Numpy Arrays

To start off, we can make 1-d Numpy arrays from non-nested lists/tuples/anything which is list-like.

```
test_list = [1,2,3,4]

test_array_1d = np.array(test_list)

#We can now access the np.array attribute, 'shape' which returns the dimensions of the array. In this case, it is a 1-d array with 4 entries (rows).
test_array_1d.shape
```

We can also make 2-d Numpy arrays

```
test_list_of_lists = [[1,2],[3,4],[5,6],[7,8]] #list of lists which can be thought of like matrices

test_array_2d = np.array(test_list_of_lists)

test_array_2d.shape #this array has 2 dimensions, and has 4 rows and 2 columns.
```

Here are some useful numpy functions to create certain special numpy arrays.

```
np.ones(shape = 6) #creates a numpy array of shape (6,)
```

```
np.zeros(shape = (5,3)) #creates a numpy array of shape (5,3)
```

```
np.random.normal(0, 1, size = (2, 3)) #creates a numpy array of shape (2,3) of normal random variables, with mean 0 and standard deviation 1
```

```
np.arange(0, 20, 3) #creates a numpy 1-d vector of all elements from 0 (inclusive) to 20 (exclusive), by jumps of 3
```

## 1.2 Pandas Basics

```
import pandas as pd
```

Pandas is a highly useful package to store your data in Python. Most notably is it's DataFrame data structure to store your datasets in an easy to view manner. If you have used R before, this is very similar to R dataframes. It is built upon Numpy arrays.

We can create a Pandas DataFrame from a dictionary (not exhaustive). 

```
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]} # dict of each column with it's values
df = pd.DataFrame(data)
df
```

Let us rename our rows (though this step is usually not necessary) for easier understanding of a later section.

```
# Rename Rows
df.rename(index={0:"first", 1:"second", 2:"third",
                 3:"fourth", 4:"fifth", 5:"sixth"}, inplace=True)
df
```

We can use the .info() method to get the number of non null entries as well as to see how each variable is being stored, the dtypes here are float64 and object.

```
df.info()
```

We can also use the .describe() method to see the summary statistics of the variables in the dataset. This is one way to see non-sensical values, for example age cannot be negative.

```
df.describe()
```

Here are some very basic functions to access values in the data

**df.iloc**
- Access data by the index, starting from 0
- df.iloc[*row(s), column(s)*]

```
print(df.iloc[1, 1])              # Accessing the value at row index 1, and column index 1
print("------")

print(df.iloc[1, :])              # Accesing the entire row at row index 1
print("------")

print(df.iloc[:, 1])              # Accesing the entire column at column index 1 (a.k.a "year")
```

**df.loc**
- Access data by their row names and column names
- df.loc[*rowname(s), colname(s)*]

```
print(df.loc['second', 'year'])   # Accessing the value at rowname 'second', and colname 'year
print("------")

print(df.loc['second', :])        # Accessing the entire row at rowname 'second'
print("------")

print(df.loc[: ,'year'])          # Accesing the entire column 'year
```

Here are some very basic splicing code to get a subset of the data

```
# Interested in obtaining only a few columns of the data
df[["state", "year"]]
```

```
df[:2] # First 2 rows
```

```
df['state'].value_counts()  # Get the frequency counts of each level for a CATEGORICAL variable
```

We can also do Boolean indexing (similar to Numpy)

```
df[df['pop'] > 2.0] # Select rows with population greater than 2.0
```

We can do some simple vectorised operations on the columns or rows as well. 

We can see the selected columns/rows as NumPy arrays, and perform the same operations that we have learnt earlier.

```
# Addition of '1.0' to the entire column
df['pop'] += 1.0
print(df)

# Addition of '1.0' to only the 'pop' of the row with row name 2.
df.loc['third', 'pop'] += 1.0
print(df)
```

# Dataset

```
#@title Download Dataset
#@markdown Run this cell to download the California Housing Dataset that we'll
#@markdown be using for today's session.

import os
import tarfile
import urllib
import pandas as pd

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    os.makedirs(housing_path, exist_ok=True)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)

fetch_housing_data()

data = load_housing_data()
```

### 3.1.1 Imputing Numerical Features

```
from sklearn.impute import SimpleImputer

# Retrieve numeric features only
data_num = data.drop('ocean_proximity', axis=1)

# Initialise Imputer for numeric features
num_imputer = SimpleImputer(strategy='median')

# Computes the median for each feature
# Performs the imputation and returns a new Numpy array
data_num_imputed = num_imputer.fit_transform(data_num)
```

**(Just FYI)** After performing the imputation on our Pandas DataFrame, our output is now a 2D Numpy array, with no headers

```
print(type(data_num_imputed))
print("-----------")
print(data_num_imputed)
```

We have performed the imputation for our numeric features, more specifically, for the "total_bedrooms" feature. Let us confirm this:

```
num_colnames = list(data_num.columns)
pd.DataFrame(data_num_imputed, columns=num_colnames).info()
```

### 3.1.2 Imputing Categorical Features

We need not impute categorical features in this case, as there are no missing values.

Nevertheless, we shall look at the code to perform the imputation

```
# Retrieve categorical features only
data_cat = data[['ocean_proximity']]

# Initialise Imputer for categorical features
cat_imputer = SimpleImputer(strategy="most_frequent", fill_value="NA")

# Performs the imputation and returns a new Numpy array
data_cat_imputed = cat_imputer.fit_transform(data_cat)
```

## 3.2 Feature Encoding

```
data['ocean_proximity'].value_counts()
```

```
from sklearn.preprocessing import OneHotEncoder

one_hot = OneHotEncoder(sparse=False, handle_unknown='ignore')

# Performs the encoding and returns a new Numpy array
data_cat_encoded = one_hot.fit_transform(data_cat_imputed)
```

```
cat_colnames = list(one_hot.categories_[0])
pd.DataFrame(data_cat_encoded, columns=cat_colnames).head()
```

## 3.3 Feature Engineering

```
# Let's combine our numerical and categorical features back into a single 
# DataFrame again
data_preprocessed_np = np.column_stack((data_num_imputed, data_cat_encoded))
data_preprocessed = pd.DataFrame(data_preprocessed_np, columns=num_colnames + cat_colnames)

data_preprocessed.head()
```

```
# Possible feature engineering: create a new rooms_per_household feature
data_preprocessed['rooms_per_household'] = (data_preprocessed['total_rooms']
                                            / data_preprocessed['households'])
```
